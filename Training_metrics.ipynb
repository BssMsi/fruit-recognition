```python
import torch
from glob import glob
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
#mpl.rcParams['text.color'], mpl.rcParams['axes.labelcolor'], mpl.rcParams['xtick.color'], mpl.rcParams['ytick.color'] = ['white']*4
%matplotlib inline
```

# Plot training and validation metrics

```python
models = glob('attempts/models/pretrained/*')
n = len(models)
```

```python
fig, ((train_acc, val_acc), (train_loss, val_loss)) = plt.subplots(2, 2, sharex=True, figsize=(15, 12))

for i in range(n):
    saved = torch.load(models[i])
    metrics = saved['metrics']
    # PLOT
    train_acc.plot(metrics['train_acc'], label=models[i].split('\\')[-1])
    val_acc.plot(metrics['val_acc'])
    train_loss.plot(metrics['train_loss'])
    val_loss.plot(metrics['val_loss'])
    
yticks = range(95, 105, 5)
train_acc.set_title('Training Accuracy')
train_acc.set_yticks(yticks)
train_acc.set_ylim([min(yticks), max(yticks)+1])
val_acc.set_title('Validation Accuracy')
val_acc.set_yticks(yticks)
val_acc.set_ylim([min(yticks), max(yticks)+1])

loss_yticks = np.arange(0, 0.5, 0.1)
train_loss.set_title('Training Loss')
train_loss.set_yticks(loss_yticks)
train_loss.set_ylim([min(loss_yticks), max(loss_yticks)])
val_loss.set_title('Validation Loss')
val_loss.set_yticks(loss_yticks)
val_loss.set_ylim([min(loss_yticks), max(loss_yticks)])

fig.legend(loc='center right');
# add a big axis, hide frame
fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.xlabel("Epochs ->")
plt.ylabel("Metric")
plt.tight_layout()
```
